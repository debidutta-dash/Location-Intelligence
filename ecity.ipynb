```python
import pandas as pd 
import gmplot
import pandas as pd
import requests
import time
```

```python
df1 = pd.read_csv('location data.csv')
df1 = df1.loc[~((df1['Category'] == 'Restaurants') | (df1['Category'] == 'Restaurant'))]
df1 = df1.reset_index(drop=True)
```

```python
time_epoch_09_30 = 1652241600        #Date and time (your time zone): Monday, May 9, 2022 9:30:00 AM
time_epoch_19_00 = 1652275800 
time_epoch_13_30= 1652256000
```

```python
e_city_centroid = (12.84526301, 77.66333012)
e_city_voosh_kitchen = (12.8418472, 77.6463297)
```

```python
from haversine import haversine


ecity_df = df1

ecity_df['haversine_distance_from_centroid'] = None
ecity_df['haversine_distance_from_voosh_kitchen'] = None
for i in range(len(ecity_df)):
  lat = ecity_df['Latitude'][i]
  lon = ecity_df['Longitude'][i]
  coord = (lat, lon)
  ecity_df['haversine_distance_from_centroid'][i] = haversine(e_city_centroid, coord)
  ecity_df['haversine_distance_from_voosh_kitchen'][i] = haversine(e_city_voosh_kitchen, coord)


ecity_df_from_centroid = ecity_df.loc[ecity_df['haversine_distance_from_centroid'] <= 2.5]
ecity_df_from_voosh = ecity_df.loc[ecity_df['haversine_distance_from_voosh_kitchen'] <= 2.5]

ecity_df_from_centroid_other = ecity_df.loc[ecity_df['haversine_distance_from_centroid'] > 2.5]
ecity_df_from_voosh_other = ecity_df.loc[ecity_df['haversine_distance_from_voosh_kitchen'] > 2.5]

print(ecity_df_from_centroid.shape)
print(ecity_df_from_voosh.shape)
print(ecity_df_from_centroid_other.shape)
print(ecity_df_from_voosh_other.shape)
```

```python
ecity_df_from_centroid=ecity_df_from_centroid.reset_index()
ecity_df_from_voosh=ecity_df_from_voosh.reset_index()
ecity_df_from_centroid_other= ecity_df_from_centroid_other.reset_index()
ecity_df_from_voosh_other=ecity_df_from_voosh_other.reset_index()
```

```python
ecity_df_from_voosh['09_30_voosh'] = None
ecity_df_from_voosh['19_00_voosh'] = None
ecity_df_from_voosh['13_30_voosh'] = None
ecity_df_from_voosh["distance_in_meters_from_voosh"] = None
```

```python
for i in range(len(ecity_df_from_voosh)):
    


  #19:00

  lat_i = ecity_df_from_voosh["Latitude"][i]
  lon_i = ecity_df_from_voosh["Longitude"][i]

  url_btm_df_from_voosh_19_00 = "https://maps.googleapis.com/maps/api/distancematrix/json?origins={}%2C{}&destinations={}%2C{}&departure_time={}&key=AIzaSyAE6DIL0N8NlCvk2El_qo2C8gFmijXRjl0".format(e_city_voosh_kitchen[0], e_city_voosh_kitchen[1], lat_i, lon_i, time_epoch_19_00)

  payload={}
  headers = {}
  response = requests.request("GET", url_btm_df_from_voosh_19_00, headers=headers, data=payload)

  ecity_df_from_voosh["19_00_voosh"][i] = response.json()['rows'][0]['elements'][0]['duration_in_traffic']['text']
  ecity_df_from_voosh["distance_in_meters_from_voosh"][i] = response.json()['rows'][0]['elements'][0]['duration_in_traffic']['value']

  #09:30


  url_btm_df_from_voosh_09_30 = "https://maps.googleapis.com/maps/api/distancematrix/json?origins={}%2C{}&destinations={}%2C{}&departure_time={}&key=AIzaSyAE6DIL0N8NlCvk2El_qo2C8gFmijXRjl0".format(e_city_voosh_kitchen[0], e_city_voosh_kitchen[1], lat_i, lon_i, time_epoch_09_30)

  payload={}
  headers = {}
  response2 = requests.request("GET", url_btm_df_from_voosh_09_30, headers=headers, data=payload)

  ecity_df_from_voosh["09_30_voosh"][i] = response2.json()['rows'][0]['elements'][0]['duration_in_traffic']['text']


  url_btm_df_from_voosh_13_30 = "https://maps.googleapis.com/maps/api/distancematrix/json?origins={}%2C{}&destinations={}%2C{}&departure_time={}&key=AIzaSyAE6DIL0N8NlCvk2El_qo2C8gFmijXRjl0".format(e_city_voosh_kitchen[0], e_city_voosh_kitchen[1], lat_i, lon_i, time_epoch_13_30)

  payload={}
  headers={}
  response3 = requests.request("GET",url_btm_df_from_voosh_13_30,headers=headers,data=payload)

  ecity_df_from_voosh["13_30_voosh"][i]= response3.json()['rows'][0]['elements'][0]['duration_in_traffic']['text']


  time.sleep(1)
```
